```r
patient.id <- "43E7BL"
cov.dir <- paste(getwd(),"out", patient.id,"erna-coverage", sep="/")

bed.to.granges <- function(bed.path){
    
    df <- read.table(bed.path, sep="\t")
    res <- GRanges(seqnames = df$V1,
                   ranges = IRanges(start = df$V2, end = df$V3),
                   read.hits = df$V4,
                   bp.hits = df$V5,
                   total.bp = df$V6,
                   pct.cov = df$V7
                  )
    return(res)
}


for (f in files){
    print(f)
    rs <- bed.to.granges(paste(cov.dir, f, sep="/"))
    print(rs[rs$pct.cov > 0,])
}
```
